```python
data = input().split()

def create_submat (match, mismatch, alphabet):
    sm = {}
    for c1 in alphabet:
        for c2 in alphabet:
            if (c1 == c2):
                sm[c1+c2] = match
            else :
                sm[c1+c2] = mismatch

    return sm

def score_pos (c1, c2, sm, g):
    if c1 == "_" or c2=="_":
        return g
    else:
        return sm[c1+c2]
def max3t (v1, v2, v3):
    if v1 > v2:
        if v1 > v3: return 1
        else : return 3
    else :
        if v2 > v3: return 2
        else : return 3
        
def smith_Waterman (seq1, seq2, sm, g):
    S = [[0]]
    T = [[0]]
    maxscore = 0
    for j in range (1, len (seq2)+1):
        S[0].append(max(0, g*j))
        T[0].append(0)
    for i in range (1, len (seq1)+1):
        S.append([max(0, g*i)])
        T.append([0])

    for i in range (0, len (seq1)):
            for j in range(len (seq2)):
                s1 = S[i][j] + score_pos (seq1[i], seq2[j], sm, g);
                s2 = S[i][j+1] + g
                s3 = S[i+1][j] + g
                b = max(s1, s2, s3)
                if b <= 0:
                    S[i+1].append(0)
                    T[i+1].append(0)
                else :
                    S[i+1].append(b)
                    T[i+1].append(max3t(s1, s2, s3))
                if b > maxscore:
                    maxscore = b
    return S, T, maxscore

def max_mat(mat):
    maxval = mat[0][0]
    maxrow = 0
    maxcol = 0
    for i in range (0, len (mat)):
        for j in range (0, len (mat[i])):
            if mat[i][j] > maxval:
                maxval = mat[i][j]
                maxrow = i
                maxcol = j

    return (maxrow,maxcol)    


def recover_align_local (S, T, seq1, seq2):
    res1 = ''
    res2 = ''
    i, j = max_mat(S)
    
    k, h = i, j
    while T[i][j]>0:
        if T[i][j]==1:
            res1 = seq1[i-1] + res1
            res2 = seq2[j-1] + res2
            i -= 1
            j -= 1
        elif T[i][j] == 3:
            res1 = "_" + res1
            res2 = seq2[j-1] + res2
            j -= 1
        elif T[i][j] == 2:
            res1 = seq1[i-1] + res1
            res2 = "_" + res2
            i -= 1
    res1 = seq1[:i].lower() + res1
    res2 = seq2[:j].lower() + res2
    res1 = res1 + seq1[k:].lower()
    res2 = res2 + seq2[h:].lower()
    return res1, res2

seq1 = data[0]
seq2 = data[1]
m = float(data[2])
mm = float(data[3])
g = float(data[4])
sm = create_submat(m, mm, 'AGCT')
S, T, maxscore = smith_Waterman (seq1, seq2, sm, g)
an1, an2 = recover_align_local (S, T, seq1, seq2)
if '_'*len(an2) in an1:
    print(maxscore, an1.upper(), (int(maxscore) - len(an2))*'_' + an2)
else:
    print(maxscore, an1, an2)
print(S)
```
