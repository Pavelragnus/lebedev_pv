## How to choose k?

**Тогда как выбрать оптимальное значение K?**  
Не существует заранее определенных статистических методов для нахождения наиболее благоприятного значения К.  
Инициализируйте случайное значение K и начните вычисления.  
Выбор малого значения К приводит к неустойчивым границам принятия решений.  
Существенное значение K лучше подходит для классификации, так как приводит к сглаживанию границ решений.  
Постройте график между частотой ошибок и значением K, обозначающим значения в определенном диапазоне. Затем выберите значение K как имеющее минимальную частоту ошибок.  
Теперь вы получите представление о выборе оптимального значения K путем реализации модели.  


- Мы получили точность 0,41 при k=37 , что выше эффективности, рассчитанной при k=4.
- Небольшое значение K не подходит для классификации.
- **Обычно оптимальное значение K представляет собой квадратный корень из N, где N — общее количество выборок.**
- Используйте график ошибок или график точности, чтобы найти наиболее благоприятное значение K.
- KNN хорошо работает с классами с несколькими метками, но вы должны знать об отклонениях.
- KNN широко используется в области распознавания образов и аналитической оценки .

```python
from scipy.spatial import distance
import numpy as np
coords = [(28, 10), # set of dots, which we've known
          (49, 49),
          (48, 35),
          (36, 33),
          (45, 54)]
coords2 = [(33,47)] # dot is new. We want to see, which are nearer to it? We can use different metrics. 
distance.cdist(coords, coords2, 'Euclidean')
```

```python
d = distance.cdist(coords, coords2, 'Euclidean')
print(d)
```

```python
weights = [1/i**2  for i in d]
```

```python
print(weights)
```

```python
Rep1 = weights[0] + weights[1] + weights[3]
```

```python
Rep2 = weights[2] + weights[4]
```
